```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from faker import Faker
fake=Faker('en_IN')
genders=['Male','Female']
education=['Graduate','Bachelors','Masters','Highschool','Diploma']
marital=['Married','Single','Divorced','Widowed']
data=[]
for _ in range(1000):
    first_name=fake.first_name()
    last_name=fake.last_name()
    name=first_name+" "+last_name
    city=fake.city()
    age=random.randint(15,60)
    gender=random.choice(genders)
    education_level=random.choice(education)
    weight=random.randint(50,100)
    marital_status=random.choice(marital)

    data.append({
        'FullName':name,
        'Gender':gender,
        'City':city,
        'EducationLevel':education_level,
        'MaritalStatus':marital_status,
        'Weight':weight,
        'Age':age,
    })
    df=pd.DataFrame(data)

df.to_csv("task1_dataset.csv", index=False)
print(df.head())

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="white")

plt.figure(figsize=(12, 7), facecolor='white')
ax = sns.histplot(df['Age'], bins=20, color='skyblue', edgecolor='black', kde=True)

ax.set_facecolor("white")  # Plot area background
plt.title('Distribution of Age', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Individuals', fontsize=12)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
plt.rcParams.update({'figure.facecolor': 'white', 'axes.facecolor': 'white'})

plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=df, color='pink')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender')
plt.grid(True)
plt.show()
```

```python
plt.rcParams.update({'axes.facecolor': 'white'})

top_cities = df['City'].value_counts(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, color='yellow')
plt.xlabel('City')
plt.ylabel('Frequency')
plt.title('Top 10 Cities')
plt.grid(True)
plt.show()
```

```python
plt.rcParams.update({'figure.facecolor': 'white', 'axes.facecolor': 'white'})

plt.figure(figsize=(8, 5))
sns.histplot(df['Weight'], bins=20, kde=True, color='salmon')
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.rcParams.update({'axes.facecolor': 'white'})
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='MaritalStatus')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

```python
plt.rcParams.update({'axes.facecolor': 'white'})
edu_counts = df['EducationLevel'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=edu_counts.index, y=edu_counts.values,color='purple')
plt.title('Education Level Distribution')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
